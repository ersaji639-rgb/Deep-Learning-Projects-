#### Neural network

input -->Hidden --> op layers

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

- **TensorFlow**: A deep learning library developed by Google.  
- **Keras**: A high-level API (wrapper) built on top of TensorFlow.  
- **Random seed**: When training neural networks in TensorFlow + Keras, certain weights are initialized randomly. Setting a random seed ensures reproducibility of results.  


```python
# Setting up the layers:
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),       # Input Layer (30 features → 1D array)
    keras.layers.Dense(20, activation='relu'),     # Hidden Layer with 20 neurons and ReLU activation
    keras.layers.Dense(2, activation='sigmoid')    # Output Layer with 2 neurons and Sigmoid activation
])
```

### Model Architecture

- **Flatten Layer**:  
  Converts the input feature matrix into a 1D array.  
  Since we have 30 features, this layer ensures each feature is passed to a corresponding neuron in the input layer.  

- **Dense Layer (Hidden Layer)**:  
  - Contains 20 neurons.  
  - Uses **ReLU (Rectified Linear Unit)** activation function:  
    - Outputs 0 for negative values, and keeps positive values as they are.  
    - Helps the model learn non-linear relationships efficiently.  

- **Dense Layer (Output Layer)**:  
  - Contains 2 neurons (for binary classification: e.g., Malignant vs Benign).  
  - Uses **Sigmoid activation function**:  
    - Maps output values between 0 and 1.  
    - Useful for probability estimation in classification tasks.  


# Compiling the Neural Network
We use the Adam optimizer, sparse categorical cross-entropy loss (since the target labels are integers), and track accuracy as the evaluation metric.


```python
model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

### One-Hot Encoding and Loss Functions

- Example of **One-Hot Encoding**:
  - Dog → [1, 0, 0]  
  - Cat → [0, 1, 0]  
  - Horse → [0, 0, 1]  

---

### Choice of Loss Function

- **Single-label classification (each sample belongs to exactly one class):**  
  - Use **Categorical Cross-Entropy Loss** (`CategoricalCrossentropy`)  

- **Multi-label classification (a sample can belong to multiple classes at the same time):**  
  - Use **Binary Cross-Entropy Loss** (`BinaryCrossentropy`)  


### Training the NN

### Explanation of `model.fit(X_train, Y_train, validation_split=0.1, epochs=10)`

This function is used to **train the neural network model**.  
It takes the training data, iteratively updates the model’s weights, and improves prediction performance.

- **X_train**  
  The input features (images, text, numerical data, etc.) that the model uses to learn patterns.

- **Y_train**  
  The target/output labels corresponding to each training example in `X_train`.

- **validation_split=0.1**  
  A fraction of the training data (here **10%**) is reserved for validation.  
  - Training set: 90% of the data  
  - Validation set: 10% of the data  
  This helps monitor how well the model generalizes to unseen data during training and prevents **overfitting**.

- **epochs=10**  
  The number of times the model goes through the **entire training dataset**.  
  Example: If there are 50,000 training samples, one epoch means the model sees all 50,000 once. After 10 epochs, the model will have seen the dataset 10 times.

---

### **Overall Meaning**  
Train the model for **10 epochs** using `X_train` and `Y_train`, while holding out **10% of the training data** for validation at the end of each epoch to check model performance.


```python
history = model.fit(X_train_standard, Y_train, validation_split=0.1, epochs= 15)
```

#### Visualizing the accuracy and loss:

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title("NN-Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Training Data", "Validation Data"], loc="lower right")
plt.show()
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title("NN-Model loss")
plt.ylabel("Aloss")
plt.xlabel("Epoch")
plt.legend(["Training Data", "Validation Data"], loc="upper right")
plt.show()
```

####  Accuracy of the model on test data:

```python
loss, accuracy = model.evaluate(X_test_standard,Y_test)
print(accuracy)
```

#### Building the predictive system:

```python


input_data = tuple(df.iloc[1, :-1])

# Change the input data to a numpy array 
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpyarray as we are predicting for 1 data point 
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# Standardize the input reshaped data 
input_data_Standardize = scaler.transform(input_data_reshape)

# Prediction 
prediction = model.predict(input_data_Standardize)

print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0]  == 0):
    print("Malignant")
else:
    print("Benigin")
```
